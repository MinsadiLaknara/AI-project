Even among children, diabetes is a condition that is rapidly increasing in prevalence. We must comprehend what occurs in the body without diabetes if we are to comprehend diabetes and how it develops. Sugar (glucose) is derived from the meals we eat, notably those high in carbohydrates. Everyone, including those with diabetes, needs carbs because they are the body's primary source of energy. Bread, cereal, pasta, rice, fruit, dairy products, and vegetables are examples of foods high in carbohydrates (especially starchy vegetables). These foods are converted into glucose by the body when we eat them. In the bloodstream, glucose circulates throughout the body.
To assist in our ability to think clearly and perform, some of the glucose is transported to our brain. The remaining glucose is transferred to our body's cells for usage as fuel, and it is also stored as energy in our liver for later use by the body. Insulin is necessary for the body to utilize glucose for energy. The beta cells in the pancreas create the hormone insulin. Insulin functions as a door's key. In order to allow glucose to enter the cell from the blood stream, insulin attaches to the cell's doors, opening them.
Glucose builds up in the bloodstream (hyperglycemia) and diabetes occurs if the pancreas is unable to generate enough insulin (insulin deficit) or if the body is unable to utilize the insulin it produces (insulin resistance). Diabetes Mellitus is characterized by elevated glucose (sugar) levels in the blood and urine.
When a person has type 1 diabetes, their immune system is weakened and their cells are unable to make enough insulin. There are no convincing studies that demonstrate the causes of type 1 diabetes, and there are also no effective preventative measures at this time.
Type 2 diabetes is characterized by either insufficient insulin production by the cells or improper insulin use by the body. 90% of people with diabetes have this kind of diabetes, making it the most prevalent type. Both genetic and lifestyle factors contribute to its occurrence.

Pregnant women who acquire high blood sugar quickly get gestational diabetes. It will recur in two-thirds of the instances during consecutive pregnancies. Following a pregnancy in which gestational diabetes was present, there is a high likelihood that type 1 or type 2 diabetes may develop. Symptoms of Diabetes

### Random forest 

A combination of tree predictors makes up the flexible, quick, and easy machine learning technique known as the random forest approach. Most of the time, random forest produces outcomes that are satisfactory. Its performance is difficult to enhance, and it can deal with a variety of data kinds, including numerical, binary, and nominal data. To produce more appropriate and precise outcomes, random forest constructs many decision trees and aggregates them. Both classification and regression have made use of it. Machine learning's main duty is classification. Similar hyperparameters to those of a decision tree or bagging classifier are present. It is simple to investigate the fact that random trees overlap in random forests. 
Let's say that, of the seven random trees that offered the information connected to some variable, four of them agree while the other three don't. The machine learning model is built based on probabilities and majority vote. In random forest, fixing a random threshold for all attributes rather than locating the best accurate threshold leads in more random trees and more accurate results on large datasets. The overfitting problem is also resolved by this approach.
--->	The diabetes data set consists of 2000 data points, with 9 features each.
--->	Outcome” is the feature we are going to predict, 0 means No diabetes, 1 means diabetes.

Advantages

1.Random Forest can be used to address problems involving classification and regression.

2.Random Forest is effective with both continuous and categorical variables.

3.Missing values can be handled easily by Random Forest.

4.No feature scaling is necessary. In the case of Random Forest, no feature scaling (standardization and normalization) is necessary because it employs a rule-based methodology as opposed to distance calculation.

5.Effectively manages non-linear parameters: Unlike curve-based algorithms, the performance of a Random Forest is unaffected by non-linear factors. Therefore, Random Forest may perform better than other curve-based algorithms if there is significant nonlinearity between the independent variables.

Disadvantages 

1.Complexity: Unlike decision trees, which only generate one tree, random forests create several trees and integrate their results. It builds 100 trees in the Python Sklearn module by default. This algorithm needs a lot more resources and computing power to accomplish this. The decision tree, on the other hand, is straightforward and doesn't need as much computing power.

2.Longer Training Period: Because Random Forest generates several trees (instead of just one tree, as with decision trees), and because it bases decisions on the majority of votes, it takes significantly longer to train than decision trees.


### Gradient Boost 

Gradient boosting has been used in a variety of technological domains over the years. The algorithm may appear complex at first, but in the majority of cases, we just utilize one standard configuration for classification and one for regression. Of course, these configurations can be changed to meet your specific needs. A family of potent machine-learning methods known as gradient boosting machines has achieved notable success in a variety of real-world applications. They can be learned with regard to various loss functions, for example, and are extremely adaptable to the specific needs of the application. The concept of gradient boosting approaches is introduced in this article with a significant emphasis on modelling aspects of machine learning. Descriptive examples and illustrations that span all the steps of the gradient boosting model design are used to supplement theoretical material. It is explained how to handle the complexity of the model. We show and thoroughly examine three real-world applications of gradient boosting.

Three factors are involved in gradient boosting:

1.A loss function that has to be improved.
2.A poor predictor of outcomes.
3.A model that adds weak learners in order to reduce the loss function


Advantages

1.Often has remarkable forecasting accuracy.

2.Great flexibility - the function fit has a wide range of hyper parameter tuning possibilities and may be optimized on various loss functions.

3.No pre-processing of the data is necessary; category and numerical values frequently perform well when presented as-is.

4.Deals with missing data; imputation is not necessary.


Disadvantages

1.Boosting of gradients Models will keep getting better to reduce all inaccuracies. This may lead to overfitting and an overemphasis on outliers.

2.Expensive to compute since it frequently requires multiple trees (more than 1000), which can be memory and time consuming.

3.The high degree of flexibility leads to a lot of variables that interact and have a significant impact on how the strategy behaves (number of iterations, tree depth, regularization parameters, etc.). This necessitates a thorough grid search while tweaking.


### Artificial Neural Network (ANN)

An important data mining tool is the artificial neural network, a field of artificial intelligence research. The input, hidden, and output layers are the three layers of an ANN. Units that translate the input layer into the output layer make up the hidden layer. One neuron's output serves as an additional layer's input. An artificial neural network recognizes complicated patterns and learns from them. There are billions of neurons in the human brain. A single neuron in this group is referred to as a perceptron, and these cells are linked to other cells by axons. Dendrites accept input, which they interpret as stimuli. Similar to this, the ANN is made up of numerous nodes connected to one another.
The link between two units is symbolized by a weight. The goal of an ANN is to transform input into meaningful output. The term "input" refers to the combination of a group of input values connected to a weight vector, where the weight may be either positive or negative. Both supervised and unsupervised learning methods are compatible with ANN. Because the model receives the output, supervised learning was used in our investigation. Input and output are both known in supervised learning. The actual output is compared to the desired outputs following processing. After then, errors are returned to the system for correction. The input is processed repeatedly during training so that the network can fine-tune and alter the weights.

Advantages 

1.Information is saved on the entire network rather than in a database, as is the case with traditional programming. The network continues to operate even if a few pieces of information in one location disappear.

2.Capacity to operate under uncertain conditions: Following ANN training, the data may still produce results under uncertain circumstances. The degree to which the performance is lost here depends on how critical the missing data is.

3.Tolerant of errors: An ANN that has one or more corrupted cells can nevertheless generate output. The networks are fault resilient because of this feature.

Disadvantages

1.Hardware dependence: Due to the topology of artificial neural networks, parallel processors are a requirement. The equipment's actualization is therefore dependent on this.

2.The most significant issue with ANNs is their unexplained behaviors. When ANN comes up with a probing solution, it doesn't explain why or how. This makes the network less trustworthy.

3.The network's length is unknown: When the network's error on the sample is decreased to a predetermined level, training is said to be finished. We are not getting the best outcomes with this value.

```python
y = Diabetes['Outcome']


X = Diabetes.drop(['Outcome'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]

max_depth = [2,4]

min_samples_split = [2, 5]

min_samples_leaf = [1, 2]

bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf_Model = RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_RandomGrid = RandomizedSearchCV(estimator = rf_Model, param_distributions = param_grid, cv = 10, verbose=2, n_jobs = 4)
```

```python
rf_RandomGrid.fit(X_train, y_train)
```

```python
rf_RandomGrid.best_params_
```

```python
print (f'Train Accuracy - : {rf_RandomGrid.score(X_train,y_train):.3f}')
print (f'Test Accuracy - : {rf_RandomGrid.score(X_test,y_test):.3f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
Model = RandomForestClassifier(n_estimators = 41,
 min_samples_split = 2,
 min_samples_leaf = 2,
 max_depth = 4,
 bootstrap = True)
Model.fit(X_train,y_train)
a = Model.predict(X_test)

print('The accuracy :', metrics.accuracy_score(a,y_test))
```

```python
# Accuracy Results
print('The accuracy :',metrics.accuracy_score(a,y_test))
print('\n \n Confusion matrix: ', metrics.confusion_matrix(a, y_test))
print('\n\n Classification report:', metrics.classification_report(a, y_test))
```
